```python
import numpy as np
import matplotlib.pyplot as plt
from collections import defaultdict

with open('./data/diffusion_boundary_movement.txt', 'r') as f:
    line = f.readline()
    rotation = 0
    
    #from frame18
    #Calium in water 1312 um^2/s     https://www.ncbi.nlm.nih.gov/pmc/articles/PMC1435176/#:~:text=We%20report%20here%20that%20Ca2%2B,)%20cm2%2Fs%20for%20Ba.
    #GFP diffuses at around 90 micron2/sec in water at 20 C.
    #In 1 sec it diffused by roughly 9.5 microns.
    lines_frames = []
    while line:
        line = line.split()
        line_int = []
        for item in line:
            item = float(item)
            line_int.append(int(item))
        lines_frames.append(line_int)

#         x = np.array(range(0, len(line)))
#         y = np.array(line_int)/(4.667/3.2)
        
#         p = np.polyfit(x, y, 1)
#         line_x = np.linspace(x[0], x[-1], 100)
#         line_y = line_x * p[0] + p[1]    #already calibrated 
#         print(p)
#         plt.plot(line_x, line_y, c='red')
#         plt.xlabel("time (s)")
#         plt.ylabel("r  (um)")
#         plt.plot(x, y)
#         plt.title("Orientation: {}".format(rotation))
#         plt.show()
        
        line = f.readline()
        rotation += 1
    lines_frames = np.array(lines_frames)/(4.667/3.2)    ## 4.677 is the convertion for magnitude 20X.  3.2 is 
    lines_frames = np.mean(lines_frames, axis=0)
    x = np.array(range(0, int(lines_frames.shape[0])))
    plt.plot(x, lines_frames)
    plt.title("Averaged")
    plt.xlabel("time (s)")
    plt.ylabel("r  (um)")
    plt.show()
    
```

```python
velocity =[]
for i in range(1,102):
    velocity.append(lines_frames[i] -lines_frames[i-1])
    i +=1

x = np.array(range(0, int(len(velocity))))
plt.scatter(lines_frames[0:101], velocity)
plt.title("Averaged diffusion velocity ")
plt.xlabel("r (um)")
plt.ylabel("velocity (um)")
plt.show()
```

```python
mag_file = '../data/velocity-magnitude-63_63.txt'
x_coord_file = '../data/x-63_63.txt'
y_coord_file = '../data/y-63_63.txt'

mags = np.zeros((63, 63))
xs = np.zeros((63, 63))
ys = np.zeros((63, 63))

# load data
with open(mag_file, 'r') as f:
    line = f.readline().split()
    idx_r = 0
    
    while line:
        for idx_c, item in enumerate(line):
            mags[idx_r, idx_c] = item
        
        line = f.readline().split()
        idx_r += 1

with open(x_coord_file, 'r') as f:
    line = f.readline().split()
    idx_r = 0
    
    while line:
        for idx_c, item in enumerate(line):
            xs[idx_r, idx_c] = item
        
        line = f.readline().split()
        idx_r += 1

with open(y_coord_file, 'r') as f:
    line = f.readline().split()
    idx_r = 0
    
    while line:
        for idx_c, item in enumerate(line):
            ys[idx_r, idx_c] = item
        
        line = f.readline().split()
        idx_r += 1

# find the valley
# center_r, center_c = 31, 31
r_idx, c_idx = 26, 33
center_r = xs[r_idx, c_idx]
center_c = ys[r_idx, c_idx]

min_value = np

table_dist_mag = defaultdict(list)
for r in range(63):
    for c in range(63):
        if r == center_r and c == center_c:
            continue
        
        r_curr = xs[r, c]
        c_curr = ys[r, c]
        
        dist = np.sqrt((center_r - r_curr)**2 + (center_c - c_curr)**2)
        mag = mags[r, c]
        
        table_dist_mag[dist].append(mag)

# smooth, v1: dump those distances who have number of pixels less than threshold
thre = 15
table_dist_mag_smoothed = defaultdict(list)
for key in table_dist_mag:
    if len(table_dist_mag[key]) > thre:
        table_dist_mag_smoothed[key] = table_dist_mag[key]
        
# average pixels with the same distance
table_dist_mag_ave = defaultdict(float)
for key in table_dist_mag_smoothed:
    table_dist_mag_ave[key] = np.mean(np.array(table_dist_mag_smoothed[key]))

# dump those pixels whose distance larger than 400 pixels
table_dist_mag_400 = defaultdict(float)
for key in table_dist_mag_ave:
    if key < 500:
        table_dist_mag_400[key] = table_dist_mag_ave[key]

# plot the data
dist_mag = table_dist_mag_400.items()
dist_mag = sorted(dist_mag)
x_data = []
y_data = []
for dist, mag in dist_mag:
    x_data.append(dist * 0.21428571)
    y_data.append(mag)


# plt.plot(x_data, y_data, c ="blue")
y_converted_direction_data = [element * (-1) for element in y_data]
plt.scatter(x_data, y_data, c ="blue", s=10)

# plt.scatter(lines_frames[0:101]-60, velocity,c="black")



# fit a approximate line
p = np.polyfit(x_data, y_data, 1)
line_x = np.linspace(x_data[0], x_data[-1], 100)
line_y = line_x * p[0] + p[1]    #already calibrated 
plt.ylabel("Contracted velocity (um/s)")
plt.xlabel("r -r0 (um)")
plt.plot(line_x, line_y, c='red')
plt.show()
```
